```python
import numpy as np
import pandas as pd
import mdtraj as md
import matplotlib
#%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
dff = pd.read_csv('bcset6/dc_bcsim6.dat', delim_whitespace=True, header=None, names=['residue', 'chainA', 'chainB'])
dff['avg'] = (dff['chainA']+dff['chainB'])/2

dfref = pd.read_csv('dc_exp.dat', delim_whitespace=True, header=None, names=['residue','avg'] )
```

```python
root = format(math.sqrt(((dff['avg']-dfref['avg'])*(dff['avg']-dfref['avg'])).mean()), ".4f")
print("Root Mean Square Error:", root)
#format(number, ".2f")
#MSE.mean()
```

```python
## Root mean square error
MMSE = np.square(np.subtract(dff['avg'], dfref['avg'])).mean()
RMSSE = math.sqrt(MMSE)
```

```python
fig = plt.figure(1034, figsize=(10,6))
ax = fig.add_subplot(111)
ax.plot(dff['residue'][:2], dff['avg'][:2], marker='o', color='b')
ax.plot(dff['residue'][2:], dff['avg'][2:], marker='o', color='b', label='sim')
ax.plot(dfref['residue'][:2], dfref['avg'][:2], marker='o', color='r')
ax.plot(dfref['residue'][2:], dfref['avg'][2:], marker='o', color='r', label='exp')

#ax.plot(dfreff['residue'][:2], dfreff['avg'][:2], marker='o', color='g')
#ax.plot(dfreff['residue'][2:], dfreff['avg'][2:], marker='o', color='g', label='expzh')

ax.set_xlabel('# Residue', weight='bold',fontsize=24, labelpad=8)
ax.set_ylabel('Dipolar coupling (kHz)', weight='bold',fontsize=24)
ax.set_xlim([12, 37])
ax.set_ylim([6.5, 11])
ax.xaxis.set_minor_locator(plt.MultipleLocator(1))
#ax.set_yticks(np.arange(0, 400.001, 100))
ax.yaxis.set_minor_locator(plt.MultipleLocator(2))
#ax.set_yticks(np.arange(-22, 22, 2))
ax.set_xticks(np.arange(12, 37, 2))
ax.tick_params(axis ='both', which= 'major', labelsize=22, pad=4)
plt.title("DC sim6", color = 'red')
fig.savefig('Epro.DC.bc.sim6.png',bbox_inches='tight', dpi=600)
plt.legend(fontsize=20)
```
